**MAXIMUM LIKELIHOOD ESTIMATION**

```python
import pandas as pd
import numpy as np

file_path = r'/content/MLE data.xlsx'
df = pd.read_excel(file_path)

print("Dataset preview:")
print(df)

df['probability'] = df.iloc[:, 1]

df["probability"] = df["probability"].astype(str)
df["probability"] = df["probability"].str.replace('0', '1-p', regex=False)
df["probability"] = df["probability"].str.replace('1', 'p', regex=False)

print("Updated dataset with 'probability' column:")
print(df)

def log_likelihood_function():
    n = df['infected'].sum()
    m = len(df['infected'])
    p = n / m

    log_likelihood = n * np.log(p) + (m - n) * np.log(1 - p)
    print(f"\nlog_likelihood:{log_likelihood}")

    print(f"\nParameter that maximize the likelihood [p]:{p}")

log_likelihood_function()
```
